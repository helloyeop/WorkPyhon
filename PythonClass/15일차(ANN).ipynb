```python
%matplotlib inline
import matplotlib.pyplot as plt
import mglearn  # 학습도움을 위해 만든 패키지
# MLP multi layer perceptron  : FFNN(Feed forward neural network) 
# forward propagation(순전파 - 예측 분류과정(가중치가 random하게 초기화) )
# backward propagation(역전파 - 가중치 학습과정(cost function 기울기, learning rate))
# solver : 미분, learning rate 조절, 가중치 조절
from sklearn.neural_network import MLPClassifier # or MLPRegressor
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
X,y = make_moons(n_samples = 100, noise = 0.25, random_state = 3)
```

```python
# stratify : 층화 변수 선택법 (y값의 비율을 0.75:0.25로 split) ==> 무작정 random이 아님
X_train , X_test ,y_train, y_test = train_test_split(X,y, stratify = y, random_state= 42)

# multi layer => XOR 문제를 해결, 레이어가 많으면 정밀도가 자동으로 높아짐
mlp = MLPClassifier(random_state= 0).fit(X_train, y_train)
#==> 가중치 완성  -- 레이어 지정 자동으로 MLPClassifier가 가중치 레이어 처리

mglearn.plots.plot_2d_separator(mlp, X_train,fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel('특성 0')
plt.ylabel('특성 1')
```

```python
mlp.get_params()  # mlp의 매개변수
#activation : relu
# ==> 0 이하를 제거한 활성화 함수( activation function )
# 속도가 빨라짐

# solver : adam
# ==> learning-rate를 조절, momentum을 사용 
```

```python
# 레이어가 10개 이상주면 기울기 소실 문제
# 멀티레이어 : XOR문제 해결, 정확도 상승

mlp = MLPClassifier(solver='lbfgs',
                   random_state = 0, hidden_layer_sizes=[10,5,10,2]) # 10,5
# 히든 레이어 사이즈 지정 시 , 출력 차수만 지정해주면 됨(추출할 특징 수)
mlp.fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp, X_train,fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel('특성 0')
plt.ylabel('특성 1')
```

```python
mlp.coefs_

# 2 x 5, bias : 10개
#변수는  2x5, 10x5 # 10으로 나감
# 레이어 사이즈를 지정 출력차수만 지정(5,10) : 입력데이터
# 2x5( 2: 외부에서 오는 데이터 , 5 : hidden layer size), 10x5
```

```python
fig , axes = plt.subplots(2, 4, figsize = (20,8))
for axx, n_hidden_nodes in zip(axes, [10,100]):
    for ax, alpha in zip(axx, [0.1, 0.001, 0.0001, 0.00001]):
        mlp = MLPClassifier(solver = 'lbfgs', random_state= 0,
                           activation = 'tanh',
                           hidden_layer_sizes = [n_hidden_nodes, n_hidden_nodes], alpha=alpha)
        mlp.fit(X_train,y_train)
        mglearn.plots.plot_2d_separator(mlp,X_train, fill =True, alpha=.3, ax= ax)

        mglearn.discrete_scatter(X_train[:,0],X_train[:,1], y_train, ax =ax)
        ax.set_title('n_hidden=[{}, {}]\nalpha={:.4f}'.format(n_hidden_nodes, n_hidden_nodes, alpha))
        
        
# ==> hidden layer size가 10개와 100개 비교
# - 10개보다 100개일 때 곡선이 많아짐 ==> 다차원 차수
# - alpha값 : 규제값(regulazation) ==> 규제가 많아질 수록 과적합됨

# ==> 다차방정식에 의한 비선형 회귀( 다항회귀 )
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
print("유방암 데이터의 특성별 최대값 : \n{}".format(cancer.data.max(axis=0)))
cancer.feature_names
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

mlp = MLPClassifier(random_state = 42)
mlp.fit(X_train,y_train)  # 가중치 결정
print("훈련 세트 정확도 : {:.2f}".format(mlp.score(X_train,y_train)))  # 94
print("테스트 세트 정확도 : {:.2f}".format(mlp.score(X_test,y_test)))  # 92
mlp
```

```python
mean_on_train = X_train.mean(axis = 0)
std_on_train = X_train.std(axis=0)

# Z점수 표준화
X_train_scaled = (X_train - mean_on_train) / std_on_train  
X_test_scaled = (X_test - mean_on_train) / std_on_train

mlp = MLPClassifier(random_state = 0)
mlp.fit(X_train_scaled, y_train)

print("훈련 세트 정확도: {:.3f}".format(mlp.score(X_train_scaled,y_train)))
print("테스트 세트 정확도: {:.3f}".format(mlp.score(X_test_scaled,y_test)))
```

```python
mlp = MLPClassifier(max_iter = 1000, random_state= 0)   # max_iter : 1000번 반복
mlp.fit(X_train_scaled, y_train)

print("훈련 세트 정확도: {:.3f}".format(mlp.score(X_train_scaled,y_train)))
print("테스트 세트 정확도: {:.3f}".format(mlp.score(X_test_scaled,y_test)))
# 훈련, 테스트 정확도가 차이남 ==> 과적합?
```

```python
mlp = MLPClassifier(max_iter = 1000, alpha = 1, random_state= 0)   # alpha : 규제  
mlp.fit(X_train_scaled, y_train)

print("훈련 세트 정확도: {:.3f}".format(mlp.score(X_train_scaled,y_train)))
print("테스트 세트 정확도: {:.3f}".format(mlp.score(X_test_scaled,y_test)))

# ==> 훈련과 테스트 정확도가 가장 유사함 ==> 일반화된 모델
```

```python
import pandas as pd
wine = pd.read_csv('./wine_data.csv', names = ["Cultivator", "Alchol", "Malic_Acid", "Ash",
                                             "Alcalinity_of_Ash", "Magnesium", "Total_phenols",
                                             "Falvanoids", "Nonflavanoid_phenols",
                                             "Proanthocyanins", "Color_intensity", "Hue",
                                             "OD280", "Proline"], encoding="utf-8") 
wine.head()
wine.shape
```

```python
mlp = MLPClassifier(max_iter = 30 ,alpha=.1, random_state=42)


X_train, X_test ,y_train, y_test = train_test_split(wine.iloc[:,1:], wine.Cultivator, random_state = 0)
mean_on_train = X_train.mean(axis = 0)
std_on_train = X_train.std(axis=0)

# Z점수 표준화
X_train_scaled = (X_train - mean_on_train) / std_on_train  
X_test_scaled = (X_test - mean_on_train) / std_on_train
mlp.fit(X_train_scaled,y_train)
print(mlp.score(X_train_scaled,y_train))
print(mlp.score(X_test_scaled,y_test))

###############################################################################################################

X = wine.drop('Cultivator', axis = 1)
y = wine['Cultivator']

X_train , X_test, y_train, y_test = train_test_split(X,y)
print(y.unique())
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# StandardScaler (copy=True, with_mean = True, with_std = True)  # copy_ True :  원본은 유지
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

mlp=  MLPClassifier(hidden_layer_sizes = (30,30,30))  # hidden_layers_size default : 100
mlp.fit(X_train,y_train)


from sklearn.metrics import classification_report  # 정밀도, 재현율, f1-score: 2*(정밀도*재현율) / (정밀도 + 재현율)
# 정밀도 : 예측을 중심으로 생각   : TP/(TP+FP)
# 재현율 : 실제값을 중심으로 생각 : TP/(TP+TN)
from sklearn.metrics import confusion_matrix
predictions = mlp.predict(X_test)

print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
print(predictions)
print(y_test)
```

```python
# NMF( Non-negative matrix factorization ) : 비음수 행렬 분해
# - PCA는 음수와 양수의 차이를 상계해서 처리함
# - NFM는 양수인 데이터에 적용 : 음성데이터, signal 데이터
```

```python
S = mglearn.datasets.make_signals()
plt.figure(figsize = (6,1))
plt.plot(S, '-')
plt.xlabel('시간')
plt.ylabel('신호')
plt.margins(0)
```

```python
# 노이즈 : 전송 데이터
import numpy as np

A = np.random.RandomState(0).uniform(size = (100,3))
X= np.dot(S,A.T)
print("측정 데이터 형태: {}".format(X.shape))
```

```python
from sklearn.decomposition import NMF, PCA
nmf = NMF(n_components = 3, random_state = 42)
S_ = nmf.fit_transform(X)
print("복원한 신호 데이터 형태 : {}".format(S_.shape))

plt.plot(S_,'-')
plt.margins(0)
```

```python
plt.plot(X,'-')
plt.margins(0)
```

```python
pca = PCA(n_components = 3)
H= pca.fit_transform(X)
plt.plot(H,'-')
plt.margins(0)
```
